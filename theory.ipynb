```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot 5 points in 2D
x = np.array([4, 4, 4, 8, 12])
y = np.array([1, 5, 10, 1, 1])

plt.scatter(x, y, color='blue', marker='o', s=100)
# put grid dotted with color black and width 1 and some opacity
plt.grid(color='black', linestyle=':', linewidth=1, alpha=0.5)
# show 1,5,10 on y axis and 4,8,12 on x axis
plt.xticks([4, 8, 12])
plt.yticks([1, 5, 10])
# name axis with latex, and put names on top and right and both rotated by 90 degrees x is l_max and y is \mathcal{I}
# both x and y should be on max values of the axis
plt.xlabel('$\ell_{max}$', fontsize=20, labelpad=20)
# we need y to be \mathcal{I} x 10^3 but symbol instead of x we use cross
plt.ylabel(r'$|\mathcal{I}| \times 10^3$', fontsize=20, labelpad=30, rotation=0)
# also plote in same figure a y = x^2/14.4 line
x = np.linspace(0, 12, 100)
y = x**0.5 * 4
plt.plot(x, y, color='red', linestyle='-', linewidth=2)

# put latex in text inside the plot
x1 = 12
y1 = 15
plt.text(x1, y1, r'$\epsilon (|\mathcal{I}|)$', fontsize=20,color='red')
x2 = 15
y2 = 10
plt.text(x2, y2, r'$\epsilon (\ell_{max})$', fontsize=20, color='green')


# print also y = x^2 with x in [0,15]
x = np.linspace(0, 15, 100)
n = 25
y = x**2 / n
plt.plot(x, y, color='green', linestyle='-', linewidth=2)

# dont stop y in 10 but in 15
plt.ylim(0, 20)
plt.xlim(0, 20)
# show arrows in x and y axis ends
plt.arrow(0, 0, 20, 0, head_width=0.5, head_length=0.5, fc='k', ec='k')

plt.show()
```
