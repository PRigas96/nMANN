```python
import numpy as np
```

```python
# y = sin(x1) + cos(x2) , x1 in [0,pi/4], x2 in [pi/4,pi/2]
# plot for x in [0,pi/4] and [pi/4,pi/2]

y = lambda x: np.sin(x[0]) + np.cos(x[1])
x = np.array([0, np.pi/4])
x1 = np.linspace(0, np.pi/4, 100)
x2 = np.linspace(np.pi/4, np.pi/2, 100)
y1 = np.sin(x1)
y2 = np.cos(x2)

import matplotlib.pyplot as plt
plt.plot(x1, y1, label='sin(x1)')
plt.plot(x2, y2, label='cos(x2)')
plt.legend()
plt.show()






    
```
